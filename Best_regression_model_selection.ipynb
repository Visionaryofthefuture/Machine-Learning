```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
df = pd.DataFrame(pd.read_csv("/content/Data.csv"))
```

```python
df
```

```python
X = df.iloc[:, :-1].values
Y = df.iloc[: , -1].values
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X[:,3], Y)
```

Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
regressor.predict(X_test)
```

```python
Y_test
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X[:,1], Y, color = "red")
plt.plot(X[:,1], regressor.predict(X), color = "blue")
```

```python
r2_score(Y, regressor.predict(X))
```

Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_reg = DecisionTreeRegressor(random_state = 0)
tree_reg.fit(X_train, Y_train)
```

```python
Y_train
```

```python
tree_reg.predict(X_train)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X[:,2], Y, color = "red")
plt.plot(X[:, 2], regressor.predict(X), color = "blue")
```

```python
r2_score(Y, tree_reg.predict(X))
```

Random Forest Regression


```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_forest = RandomForestRegressor(random_state = 0)
```

```python
random_forest.fit(X_train, Y_train)
```

```python
random_forest.predict(X_train)
```

```python
Y_train
```

```python
r2_score(Y, random_forest.predict(X))
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 2)
```

```python
X_poly = poly_reg.fit_transform(X_train)
regressor.fit(X_poly , Y_train)
```

```python
k =regressor.predict(poly_reg.fit_transform(X_test))
```

```python
Y_test
```

```python
r2_score(Y_test, k)
```
